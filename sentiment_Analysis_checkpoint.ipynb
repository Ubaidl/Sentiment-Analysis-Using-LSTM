```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('IMDB Dataset.csv',engine="python",encoding="utf-8",on_bad_lines="skip")
```

```python
data
```

```python
data.head()
```

```python
data.shape
```

```python
data['sentiment'].value_counts()
```

```python
data.replace({"sentiment": {"positive": 1, "negative":0 }},inplace=True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Embedding,LSTM
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
x_train_data, x_test_data=train_test_split(data,test_size=0.2,random_state=42)
```

```python
x_train_data
```

```python
x_train_data.shape
```

```python
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(x_train_data['review'])
```

```python
sequences = tokenizer.texts_to_sequences(x_train_data['review'])
```

```python
padded_sequences = pad_sequences(sequences, maxlen=200)
```

```python
padded_sequences
```

```python
x_train=pad_sequences( tokenizer.texts_to_sequences(x_train_data['review']),maxlen=200)
```

```python
x_test=pad_sequences( tokenizer.texts_to_sequences(x_test_data['review']),maxlen=200)
```

```python
x_train
```

```python
x_test
```

```python
y_train = x_train_data['sentiment']
```

```python
y_train
```

```python
y_test = x_test_data['sentiment']
```

```python
y_test
```

```python
model= Sequential()
```

```python
model = Sequential()
model.add(Embedding(input_dim=5000, output_dim=128, input_length=200))
model.add(LSTM(128, dropout=0.2))
model.add(Dense(1, activation='sigmoid'))



```

```python
model.summary()
```

```python
model.compile(optimizer = "adam", loss= "binary_crossentropy", metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=5)
```

```python
model.save("model.h5")
```

```python
import joblib
joblib.dump(tokenizer,"tokenizer.pkl")
```

```python
loss ,accuracy= model.evaluate(x_test, y_test)
```

```python
print(loss)
```

```python
print(accuracy)
```

```python
#Building predicting system
```

```python
def predictive_system(review):
  sequences = tokenizer.texts_to_sequences([review])
  padded_sequences = pad_sequences(sequences,maxlen=200)
  predictions = model.predict(padded_sequences)
  sentment = "positive" if predictions[0][0]>0.5 else "negative"
  return sentment
```

```python
predictive_system("The acting was awful and the plot made no sense.")
```
